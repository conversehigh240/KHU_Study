## E 8)  [10 pts]

Use the `train_test_split` function from scikit-learn to divide the training dataset further into a training subset and a validation set. The validation set should be 20% of the training dataset size, and the training subset should be 80% of the training dataset size. 

For you reference, the `train_test_split` function is documented at http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html.

````python
from sklearn.model_selection import train_test_split


X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val,
                                                          test_size = 0.2,
                                                          train_size = 0.8,
                                                          stratify= y_train_val)
````

## E 10)  [10 pts]

Consider the we trained a 5-nearest neighbor classifier using scikit-learn on the previous training dataset:

````python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
````

````python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X_train, y_train, knn)
````

Compute *the number of* misclassifications of the 5-NN classifier on the training set. The number of errors should be a count, i.e., a positive integer.

````python
# YOUR CODE
errors = int(len(y_train) - knn.score(X_train, y_train) *len(y_train))
errors
````

## E 11)  [10 pts]

Use the code from E 10) to 

- also visualize the decision boundaries of *k*-nearest neighbor classifiers with k=3, k=7, k=9
- compute the number of misclassification on the training set, validation set and test set for the *k*-nearest neighbor classifiers with k=3, k=7, k=9

````python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
plot_decision_regions(X_train, y_train, knn)
print('Training Errors:', int(len(y_train) - knn.score(X_train, y_train) *len(y_train)))
print('Validation Errors:', int(len(y_val) - knn.score(X_val, y_val) * len(y_val)))
print('Test Errors:', int(len(y_test) - knn.score(X_test, y_test) * len(y_test)))
````

````python
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train, y_train)
plot_decision_regions(X_train, y_train, knn)
print('Training Errors:', int(len(y_train) - knn.score(X_train, y_train) *len(y_train)))
print('Validation Errors:', int(len(y_val) - knn.score(X_val, y_val) * len(y_val)))
print('Test Errors:', int(len(y_test) - knn.score(X_test, y_test) * len(y_test)))
````

## E 12)  [10 pts]

Write a for loop to evaluate different *k* nn models with k=3 to k=15. In particular, fit the `KNeighborsClassifier` on the training subset, then evaluate it on the training subset, validation subset, and test subset. Report the respective classification error or accuracy.

````python
for k in range(3, 16):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    # YOUR CODE
    train_accuracy = knn.score(X_train, y_train)*100
    val_accuracy = knn.score(X_val, y_val)*100
    test_accuracy = knn.score(X_test, y_test)*100
    print("k = ", k, f'Training accuracy = {train_accuracy:.2f}%, Validation accuracy = {val_accuracy:.2f}%, Test accuracy = {test_accuracy:.2f}%')
````
